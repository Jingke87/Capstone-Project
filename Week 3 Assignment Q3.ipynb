```python
import pandas as pd
```

```python
postal = pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')
```

```python
codes = postal[0]
indexes = codes[codes['Borough']=='Not assigned'].index
codes.drop(indexes, inplace=True)
codes['Neighbourhood'][9]="Queen's Park"
codes
```

```python
x=codes.groupby('Postcode')
new = x.aggregate(lambda p: p.unique().tolist())
new = new.reset_index()
new.rename(columns={'Postcode': 'Postal Code'}, inplace=True)
new
```

```python
new.shape
```

```python
coordinates = pd.read_csv('http://cocl.us/Geospatial_data')
coordinates
```

```python
df = pd.merge(new, coordinates, on='Postal Code', how='outer')
df
```

### Create a map of Toronto with neighborhoods superimposed on top

```python
address = 'Toronto, ON'
geolocator = Nominatim(user_agent="tn_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=10)
map_toronto
```

```python
# add markers to map
for lat, lng, borough in zip(df['Latitude'], df['Longitude'], df['Borough']):
    label = borough
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```
